```python
# Import necessary libraries
from transformers import AutoModelForCausalLM, AutoTokenizer
from trl import setup_chat_format
import torch
```

## SmolLM2 Chat Template

Let's explore how to use a chat template with the `SmolLM2` model. We'll define a simple conversation and apply the chat template.

```python
# Dynamically set the device
device = (
    "cuda"
    if torch.cuda.is_available()
    else "mps" if torch.backends.mps.is_available() else "cpu"
)

model_name = "HuggingFaceTB/SmolLM2-135M"
model = AutoModelForCausalLM.from_pretrained(
    pretrained_model_name_or_path=model_name
).to(device)
tokenizer = AutoTokenizer.from_pretrained(pretrained_model_name_or_path=model_name)
model, tokenizer = setup_chat_format(model=model, tokenizer=tokenizer)
```

```python
# Define messages for SmolLM2
messages = [
    {"role": "user", "content": "Hello, how are you?"},
    {
        "role": "assistant",
        "content": "I'm doing well, thank you! How can I assist you today?",
    },
]
```

# Apply chat template without tokenization

The tokenizer represents the conversation as a string with special tokens to describe the role of the user and the assistant.


```python
input_text = tokenizer.apply_chat_template(messages, tokenize=False)

print("Conversation with template:", input_text)
```

# Decode the conversation

Note that the conversation is represented as above but with a further assistant message.


```python
input_text = tokenizer.apply_chat_template(
    messages, tokenize=True, add_generation_prompt=True
)

print("Conversation decoded:", tokenizer.decode(token_ids=input_text))
```

# Tokenize the conversation

Of course, the tokenizer also tokenizes the conversation and special token as ids that relate to the model's vocabulary.



```python
input_text = tokenizer.apply_chat_template(messages, add_generation_prompt=True)

print("Conversation tokenized:", input_text)
```

```python
# First, load the dataset as provided
from datasets import load_dataset

ds = load_dataset("HuggingFaceTB/smoltalk", "everyday-conversations")

# Define a processing function that converts each sample into a chat format
def process_dataset(sample):
    # In this dataset, the conversation is already provided as a list of messages
    # under the "messages" key, where each message is a dictionary with "role" and "content".
    messages = sample["messages"]

    # Use the tokenizer's chat template method to convert the conversation into the required format.
    # The add_generation_prompt=True flag appends a prompt token for the upcoming assistant turn.
    chat_input_ids = tokenizer.apply_chat_template(messages, add_generation_prompt=True)

    # Optionally, store both the tokenized version and its decoded string into the sample for inspection.
    sample["chat_input_ids"] = chat_input_ids
    sample["chat_text"] = tokenizer.decode(chat_input_ids)

    return sample

# Map the processing function over the dataset
ds = ds.map(process_dataset)

# Now, for demonstration, display the processed conversation from the first training sample
print(ds["train"][0]["chat_text"])
```

```python
from IPython.core.display import display, HTML

# Display an embedded viewer for the GSM8K dataset (optional)
display(
    HTML(
        """<iframe
  src="https://huggingface.co/datasets/openai/gsm8k/embed/viewer/main/train"
  frameborder="0"
  width="100%"
  height="360px"
></iframe>
"""
    )
)

# Load the GSM8K dataset (main configuration)
from datasets import load_dataset
ds = load_dataset("openai/gsm8k", "main")

def process_dataset(sample):
    # For GSM8K, assume each sample has:
    # - "question": the problem statement (user message)
    # - "answer": the detailed solution (assistant message)

    # Create a conversation using a list of messages in the expected format:
    messages = [
        {"role": "user", "content": sample["question"]},
        {"role": "assistant", "content": sample["answer"]}
    ]

    # Apply the chat template to convert the messages into the model's required format.
    # add_generation_prompt=True appends an extra assistant prompt so the model is ready to generate.
    chat_input_ids = tokenizer.apply_chat_template(messages, add_generation_prompt=True)

    # Optionally, store both the tokenized version and the decoded text version in the sample for inspection.
    sample["chat_input_ids"] = chat_input_ids
    sample["chat_text"] = tokenizer.decode(chat_input_ids)

    return sample

# Map the processing function over the dataset
ds = ds.map(process_dataset)

# For demonstration, print the chat-formatted conversation from the first training sample
print(ds["train"][0]["chat_text"])
```
